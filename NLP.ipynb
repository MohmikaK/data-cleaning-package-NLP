```python
regular_punc = list(string.punctuation)
```

```python
extra_punc=['^',
 '?',
 '%',
 '(',
 '!',
 ')',
 ';',
 ']',
 '\\',
 '`',
 '’',
 '+',
 '~',
 '"',
 ',',
 '<',
 '‘',
 '>',
 '$',
 '}',
 '*',
 '=',
 "'",
 '|',
 '-',
 '/',
 '_',
 '{',
 '&',
 '.',
 '#',
 '”',
 '“',
 '@',
 ':',
 '[']
```

```python
all_punc = list(set(regular_punc + extra_punc))
```

```python
def remove_punctuation(text,all_punc):
    for punc in all_punc:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()
```

### Remove HTML tags


```python
import re
def remove_html(text):
    CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    text = re.sub(CLEANR,'',text)
    return text
```

```python
news1 = remove_html(news1)
news1
```

### Remove URL

```python
import re
def remove_url(text):
    tags = re.compile(r'^https?:\/\/.*[\r\n]*')
    text = re.sub(tags, '', text)
    return text
```

```python
news1 = remove_url(news1)
news1
```

### Remove Emoji

```python
import re 
def remove_emoji(text):
    emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)
    text = re.sub(emoji_pattern,'',text)
    return text
```

```python
news1 = remove_emoji(news1)
news1
```

### Remove numbers

```python
def remove_numbers(text):
    pattern = r'[0-9]'
    text = re.sub(pattern,'', text)
    return text
```
